```python
import pandas as pd
data=pd.read_csv("/content/Data-Week6 (2).csv")
data.head(2)
```

```python
x=data[["Year","Temp"]]
```

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
SSE=[]
for k in range(1,10):
    knn=KMeans(n_clusters=k)
    knn.fit(x)
    pred_i=knn.predict(x)
    SSE.append(knn.inertia_)
plt.figure(figsize=(20,10))
plt.plot(range(1,10),SSE,color='b',linestyle='dashed',marker='o',markerfacecolor='r',markersize=10)
plt.show()
```
